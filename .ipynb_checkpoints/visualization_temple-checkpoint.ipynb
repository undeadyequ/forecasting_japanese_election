```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import json
import matplotlib as mpl
import matplotlib.ticker as mtick




# global set
font = {'family' : 'Arial',
        'weight' : 'normal',
        'size'   : 16}
mpl.rc('font', **font)


csvdata = pd.read_csv("data_election.csv",delimiter=",")
csvdata = csvdata.round(2)
orig_data = csvdata.drop(['Year', 'LDP_votes', 'PM_approval2'], axis=1)
orig_data = orig_data[csvdata['Year'] != 2012]


csvdata = pd.read_csv("data_election_pred.csv",delimiter=",")
csvdata = csvdata.round(2)
pred_data = csvdata[csvdata['Year'] != 2012]

acc_data = pd.read_csv("method_acc1.csv", delimiter=",")
acc_data = acc_data.round(2)
acc_data = acc_data.set_index("Unnamed: 0")
acc_data = acc_data.reindex(index = ['lr','bag_lr','gradient_lr', "rf", "dt", "bag_dt", "adaboost_dt", "gradient_boost", "LDP_seats"])
acc_data.index.name = None

with open("feature_import.json", "r") as f:
    feats_import_dict = json.load(f)

#gdp = orig_data[:, 0]
#dur = orig_data[:, 1]
#? = orig_data[:, 2]
#seat_gd = orig_data[:, 3]
#seat_pred = pred_data[:, 4:6]

#mae_rmse = {}

methods_name_dict = {
    "lr": "LBT (2012)",
    "bag_lr": "Linear bagging",
    "gradient_lr": "Linear gradient boosting",
    "dt": "Decision tree", 
    "rf": "Random forest",
    "bag_dt": "DT-bagging",
    "adaboost_dt": "Adaboosting",
    "gradient_boost": "DT-gradient boosting",  
    "LDP_seats": "Electoral outcome \n (Ground truth)"
}

    
# Suplot of multiple-line (pred, baseline, gd)  (Fig. 4)
def shows_gd_pred(pred_data, methods=["LDP_seats", "lr", "gradient_boost"]):
    """
    1 figure
    """
    #fig, axs = plt.subplots(1, 1, figsize=(9, 3))
    #plt.figure(figsize=(40,10))
    
    #pred_data = pred_data[["Year", "LDP_seats", "lr", "gradient_boost"]]
    #pred_data = pred_data[["Year", "LDP_seats", "lr","dt","rf","bag","adaboost","gradient_boost"]]
    axs = pred_data.plot(x="Year", y=methods)
    h, l = axs.get_legend_handles_labels()
    axs.set_ylabel("LDP ratio")
    axs.figure.set_size_inches(9, 3)
    plt.legend(h, labels=[methods_name_dict[m_name] for m_name in methods], prop={'size': 12})
    plt.savefig('gd_pred.png', dpi=300, transparent=False, bbox_inches='tight')
    #axs.set_ylabel("Year")
    plt.show()

    # pred_data.plot(x="Year", y="pred_m1")
    # pred_data.plot(x="Year", y="pred_m2")

    # x = np.arange(len(seat_gd))
    # fig, ax = plt.subplots(figsize=(), layout="")
    # pred1 = ax.plot()
    # ax.plot(x, seat_gd, color="blue")
    # ax.plot(x, seat_pred, color="orange")

# suplot of multiple-line of X_Y (Fig. 1)
def show_x_y(orig_data):
    fig, axs = plt.subplots(1, 3, figsize=(9, 3))
    
    m1, b1 = np.polyfit(orig_data["GDP"], orig_data["LDP_seats"], 1)
    m2, b2 = np.polyfit(orig_data["PM_approval"], orig_data["LDP_seats"], 1)
    m3, b3 = np.polyfit(orig_data["DAYS"], orig_data["LDP_seats"], 1)
    
    orig_data["gdp_ln"] = orig_data["GDP"] * m1 + b1
    orig_data["pmapp_ln"] = orig_data["PM_approval"] * m2 + b2
    orig_data["days_ln"] = orig_data["DAYS"] * m3 + b3

    orig_data.plot(x="GDP", y="LDP_seats", kind="scatter", ax=axs[0])
    orig_data.plot(x="GDP", y="gdp_ln", kind="line", ax=axs[0], legend=False)
    axs[0].set_xlabel("GDP")
    axs[0].xaxis.set_major_formatter(mtick.PercentFormatter())
    axs[0].yaxis.set_major_formatter(mtick.PercentFormatter())
    axs[0].set_ylabel("LDP ratio")

    orig_data.plot(x="PM_approval", y="LDP_seats", kind="scatter", ax=axs[1])
    orig_data.plot(x="PM_approval", y="pmapp_ln", kind="line", ax=axs[1], legend=False)
    axs[1].xaxis.set_major_formatter(mtick.PercentFormatter())
    axs[1].set_xlabel("PM approval")
    axs[1].set_ylabel("")
    axs[1].set_xticks([20, 40, 60])
    axs[1].set_yticks([])
                                          
    orig_data.plot(x="DAYS", y="LDP_seats", kind="scatter", ax=axs[2])
    orig_data.plot(x="DAYS", y="days_ln", kind="line", ax=axs[2], legend=False)
    axs[2].set_xlabel("Days")
    axs[2].set_ylabel("")
    axs[2].set_yticks([])

                                                 
    fig.suptitle("The scatter plot of each variable and the LDP ratio")
    
    fig.tight_layout()
    plt.savefig('show_x_y.png', dpi=300, transparent=False, bbox_inches='tight')
    plt.show()


# Barchar of all performances (Fig. 2,3)
def show_bar_el_lr(acc_data, data_type="traintest", methods=["lr", "bag_lr", "gradient_lr"]):
    if data_type == "train" or data_type == "test":
        fig, axs = plt.subplots(1, 1, figsize=(8, 6))
        #plt.figure(figsize=(12, 10))
        #acc_data = acc_data[["train_mae","train_mae_var","test_mae","test_mae_var",
        #                     "train_mae","train_mae_var","test_mae","test_mae_var"]]
        acc_data = acc_data[acc_data.index.isin(methods)]
        acc_data = acc_data[[data_type + "_mae", data_type + "_rmse"]]
        print(acc_data)
        acc_data.plot(kind='barh', ax=axs)
        axs.invert_yaxis()
        #ax = acc_data.plot(kind='bar')
        x_lables = acc_data.index.values
        print(x_lables)
        axs.set_yticklabels([methods_name_dict[m_name]
                    for m_name in x_lables])
        #axs.set_yticklabels(methods)
        # the margin between bar and border
        axs.margins(x=0.4)
        
        for container in axs.containers:
            axs.bar_label(container)
        
        # legend (size)
        #axs[i].legend(loc=1,fontsize=12)
        axs.get_legend().remove()
            
        # get handler and labels
        h_i, l_i = axs.get_legend_handles_labels()        
        labels = ["MAE", "RMSE"]
        axs.legend(h_i, labels=labels,
           loc="lower right")
        
        #ax.legend(loc="best", bbox_to_anchor = (1, 1.2))
        #ax.legend(loc=1)
        plt.savefig('bar_el_lr.png', dpi=300, transparent=False, bbox_inches='tight')
        plt.show()
    else:
        fig, axs = plt.subplots(1, 2, figsize=(12, 4))
        acc_data = acc_data[acc_data["Unnamed: 0"].isin(methods)]
        acc_data_train = acc_data[["train_mae","train_rmse"]]
        acc_data_test = acc_data[["test_mae","test_rmse"]]
        
        h = []
        for i, data in enumerate([acc_data_train, acc_data_test]):
            data.plot(kind='barh', ax=axs[i])
            # subtitle
            axs[i].title.set_text("In-sample evaluation" if i == 0 else "Out-of-sample evaluation")
            axs[i].invert_yaxis()
            if i == 0:
                axs[i].set_yticklabels([methods_name_dict[m_name]
                    for m_name in methods])
                #axs[i].set_yticklabels(methods)
            else:
                axs[i].set_yticklabels([])
            # the margin between bar and border
            axs[i].margins(x=0.2)
            # show data over image
            for container in axs[i].containers:
                axs[i].bar_label(container)    
            
            # legend (size)
            #axs[i].legend(loc=1,fontsize=12)
            axs[i].get_legend().remove()
            
            # get handler and labels
            h_i, l_i = axs[i].get_legend_handles_labels()
            h.append(h_i)
        
        labels = ["MAE", "RMSE"]
        fig.legend(h, labels=labels,
           loc="upper right")
        plt.savefig('bar_el_lr.png', dpi=300, transparent=False, bbox_inches='tight')
        plt.show()

def show_bar_el_dt(acc_data, type="traintest"):
    el_lr = ["lr", "dt", "rf", "bag_dt", "adaboost_dt", "gradient_boost"]
    plt.figure()
    #acc_data = acc_data[["train_mae","train_mae_var","test_mae","test_mae_var",
    #                     "train_mae","train_mae_var","test_mae","test_mae_var"]]
    acc_data = acc_data[acc_data["Unnamed: 0"].isin(el_lr)]
    acc_data = acc_data[["test_mae","test_rmse"]]
    ax = acc_data.plot(kind="barh")
    ax.invert_yaxis()
    ax.margins(x=0.2)
    ax.set_yticklabels(["lr", "dt", "rf", "bag_dt", "adaboost_dt", "gradient_boost"])
    for container in ax.containers:
        ax.bar_label(container)  
    #ax.legend(loc=4, bbox_to_anchor = (1, 1))
    ax.legend(loc=4)
    plt.savefig('bar_el_dt.png', dpi=300, transparent=False, bbox_inches='tight')
    plt.show()


def feature_importance(feats_import_dict):
    el_lr = ["bag_dt","gradient_boost"]
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))
    data={"DT-bagging": [j for sub in feats_import_dict["bag_dt"] for j in sub],
          "DT-gradient boosting": [j for sub in feats_import_dict["gradient_boost"] for j in sub],
          "Variable": list(30 * "G") + list(30 * "P") + list(30 * "D")}
    data1={"GDP": feats_import_dict["bag_dt"][0] + feats_import_dict["gradient_boost"][0],
            "PM_approval": feats_import_dict["bag_dt"][1] + feats_import_dict["gradient_boost"][1],
            "DAYS": feats_import_dict["bag_dt"][2] + feats_import_dict["gradient_boost"][2],
            "Variable": list(30 * "A") + list(30 * "G")}

    df = pd.DataFrame(data=data)
    for i, el in enumerate(list(df.columns.values[:-1])):
        a = df.boxplot(el, by="Variable", ax=axs.flatten()[i])
        a.set_xticklabels(["Days", "GDP", "PM approval"], fontdict={"fontsize":14})
        a.set_xlabel("")
        if i==0:
            a.set_ylabel("Feature importance score", fontsize=14)
    fig.delaxes(axs[1, 0])
    fig.delaxes(axs[1, 1])
    plt.tight_layout()
    plt.suptitle("")
    plt.savefig('feature_importance.png', dpi=300, transparent=False, bbox_inches='tight')
    plt.show()
```
